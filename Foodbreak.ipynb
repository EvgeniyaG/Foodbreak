```python
#import libraries

import pandas as pd
Location =  'outbreaks.csv'
df = pd.read_csv(Location)
```

```python
df.head()
```

```python
df.describe()
```

```python
#What food caused most illnesses
pd.pivot_table(df, index=['Food'], values=['Illnesses','Hospitalizations','Fatalities' ], aggfunc='sum').sort_values('Illnesses',ascending=False)
```

```python
#In which State, when and what food caused most illnesses
pd.pivot_table(df, index=['State','Year','Month','Food'], values=['Illnesses','Hospitalizations' ], aggfunc='sum').sort_values('Illnesses', ascending=False) 
```

```python
pd_sate = pd.pivot_table(df, index=['State'], values=['Illnesses','Hospitalizations' ], aggfunc='sum')
```

```python
#In which State most illnesses happend during whole period of 1998-2015?
pd_sate.sort_values('Illnesses',ascending=False)
```

```python
#What year most illnesses happened
pd_year = pd.pivot_table(df, index=['Year'], values=['Illnesses','Hospitalizations' ], aggfunc='sum')
pd_year.sort_values('Illnesses',ascending=False)
```

```python
#Creating dataframe to find out serious illnesses which caused hospitalisations
df_hosp = df[df['Hospitalizations']>0]
```

```python
df_hosp
```

```python
#What food and Species cause most Fatalities?
pd.pivot_table(df_hosp, index=['Food','Species'], values=['Illnesses','Fatalities' ], aggfunc='sum').sort_values('Fatalities', ascending=False) 
```

```python
#Creating dataframe for Holiday season (November,December) to define what is the most common cause of illnesses
df_NY = df[(df['Month']== 'November')| (df['Month']== 'December')]
```

```python
df_NY.head()
```

```python
#What food and Species caused most fatalities during Holiday season?
pd.pivot_table(df_NY, index=['Food','Species'], values=['Illnesses','Hospitalizations','Fatalities' ], aggfunc='sum').sort_values('Fatalities', ascending=False)
```

```python
#Creating a new dataframe to select only deseases caused by turkey and chicken food during Holiday season
df_NY['Food'].fillna('Missing',inplace=True)
df_turkey_chicken = df_NY.loc[df_NY['Food'].str.contains('Turkey','Chicken')]
```

```python
#To identify Holiday food (chicken and turkey) from what place caused most illnesses during Holiday seson
df_holiday=(df_turkey_chicken.groupby('Location')['Illnesses','Hospitalizations','Fatalities'].sum())
```

```python
df_holiday.sort_values('Illnesses',ascending=False)
```

```python
tot_hosp = df_NY['Hospitalizations'].sum()
tot_hosp
```

```python
tot_ill = df_NY['Illnesses'].sum()
tot_ill
```

```python
tot_fat = df_NY['Fatalities'].sum()
tot_fat
```

```python
#Probability of hospitalization if you got illness during Holiday season
p_hosp = tot_hosp/tot_ill
p_hosp
```

```python
#Probability of fatality if you got illness during Holiday season
p_fat = tot_fat/tot_ill
p_fat
```
